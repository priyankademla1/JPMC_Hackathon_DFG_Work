```python
scaled_features = scaler.fit_transform(df_final[columns])
params = {}

for i in range(2,20):
    kmeans = KMeans(init="random",
    n_clusters=i,
     n_init=10,
    max_iter=300,
      random_state=37)
    kmeans.fit(scaled_features)
    params[i]=kmeans.inertia_
    print(f"{i} {kmeans.inertia_}")
```

```python
cluster_nums = list(params.keys())
inertia = list(params.values())
plt.scatter(cluster_nums,inertia)
plt.title("Clusters vs. Inertia")
```

```python
scaled_features = scaler.fit_transform(df_final[columns])
    
kmeans = KMeans(init="random",
    n_clusters=5,
     n_init=10,
    max_iter=300,
      random_state=37)

kmeans.fit_transform(scaled_features)
```

```python
scaled_features.shape
```

```python
centroids = kmeans.cluster_centers_
centroids.shape
```

```python
kmeans.labels_
```

```python
categories = columns
N = len(categories)
```

```python
angles = [n * 2 * pi / float(N) for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
plt.xticks(angles[:-1], categories)

# Plot each cluster
for i in range(len(centroids)):
    values = centroids[i].flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=f'Cluster {i}')
    ax.fill(angles, values, alpha=0.1)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show plot
plt.show()
```
